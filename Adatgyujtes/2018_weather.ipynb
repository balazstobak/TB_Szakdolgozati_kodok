```python
import os
import fastf1
import pandas as pd
from pathlib import Path

# Cache beállítása
cache_dir = '/content/cache'
os.makedirs(cache_dir, exist_ok=True)
fastf1.Cache.enable_cache(cache_dir)

# Ide gyűjtjük az egyes futamok időjárás adatait
all_weather_data = []

# A 2018-as szezonban 21 futam volt
for i in range(1, 22):
    print(f"{i}. futam adatainak betöltése...")
    try:
        session = fastf1.get_session(2018, i, 'R')
        
        # KIZÁRÓLAG az időjárás letöltése (drágán és lassan töltődő telemetria nélkül)
        session.load(telemetry=False, laps=False, messages=False, weather=True)
        
        # Időjárás adatok kinyerése
        weather_df = session.weather_data.copy()
        
        # A 'Race' azonosító hozzáadása (használhatod a session.event['EventName']-t is, ha a futam neve kell)
        weather_df['Race'] = i
        
        # Hozzáadjuk a listához
        all_weather_data.append(weather_df)
        
    except Exception as e:
        print(f"Hiba a(z) {i}. futam betöltésekor: {e}")

# Az összes futam adatainak egyesítése egyetlen DataFrame-be
if all_weather_data:
    final_weather_df = pd.concat(all_weather_data, ignore_index=True)

    # Mappa létrehozása és mentés
    output_dir = Path("weather")
    output_dir.mkdir(exist_ok=True)
    
    # Mentés valódi Parquet formátumban
    output_path = output_dir / "weather_2018_all.parquet"
    final_weather_df.to_parquet(output_path, index=False)
    
    print(f"\nSikeres mentés: {output_path} ({len(final_weather_df)} sor)")
else:
    print("\nNem sikerült adatokat letölteni.")
```
