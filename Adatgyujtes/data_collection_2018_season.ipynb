```python
import os
import fastf1
import numpy as np
import pandas as pd
from pathlib import Path

season = 2018

# cache
os.makedirs("cache", exist_ok=True)
fastf1.Cache.enable_cache("cache")

# output mappa
Path("laps_all_season").mkdir(exist_ok=True)

all_laps = []

for i in range(1, 22): # FUTAMSZÁM +1

    print(f"Loading Round {i}")

    # --- ÚJ RÉSZ: Kvalifikáció betöltése ---
    # Kikapcsoljuk a telemetriát, időjárást és az üzeneteket, hogy gyorsabb legyen
    quali = fastf1.get_session(season, i, "Q")
    quali.load(telemetry=False, weather=False, messages=False)
    
    # Kinyerjük az időmérős helyezéseket ('Position' oszlop a Q session-ben)
    quali_mapping = quali.results.set_index('DriverNumber')['Position'].to_dict()
    # ---------------------------------------

    # Futam betöltése a köridőkhöz
    race = fastf1.get_session(season, i, "R")
    race.load()

    laps = race.laps.copy()

    # Hozzárendeljük a kvalifikációs pozíciót a köridőkhöz
    laps['GridPosition'] = laps['DriverNumber'].map(quali_mapping)

    # metaadatok
    laps["Season"] = season
    laps["Round"] = i
    laps["EventName"] = race.event["EventName"]
    laps["Country"] = race.event["Country"]
    laps["Location"] = race.event["Location"]
    laps["EventDate"] = race.event["EventDate"]

    all_laps.append(laps)

# összefűzés
laps_df = pd.concat(all_laps, ignore_index=True)

print("Total laps:", len(laps_df))

# mentés
laps_df.to_parquet("laps_all_season/laps_2018.parquet")
```
